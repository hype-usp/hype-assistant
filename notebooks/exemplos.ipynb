```python
from dotenv import load_dotenv
import os

load_dotenv()

from llama_index.core import (
    VectorStoreIndex, 
    ServiceContext,
    set_global_service_context
)

from llama_index.llms.openai import OpenAI
from llama_index.embeddings.openai import OpenAIEmbedding
```

```python
api_key = os.getenv("OPENAI_KEY")
os.environ["OPENAI_API_KEY"] = api_key
```

```python
#####################################################################
# constants 
#####################################################################
path_prompt_folder = '../prompt/'

#####################################################################
# Models definition
#####################################################################

# llm model object
llm = OpenAI(model="gpt-3.5-turbo")

# embedding model object
embed_model = OpenAIEmbedding(
    model="text-embedding-3-small",
)

#####################################################################
# Service Context
#####################################################################
service_context = ServiceContext.from_defaults(
  llm=llm,
  embed_model=embed_model,
  system_prompt="You are an AI assistant answering questions."
)

set_global_service_context(service_context)
```

### Sistema 1

```python
while(True):

    pergunta = input()
    resposta = llm.complete(pergunta).text
    
    print('------------------------')
    print("Pergunta: ", pergunta)
    print("Resposta: ", resposta)
    print('------------------------')
```

### Sistema 2

```python
with open('../src/prompt/sistema2.txt', 'r') as file:
    prompt = file.read()
```

```python
conversation_history = []
while(True):

    pergunta = input()
    
    prompt_formated = prompt.format(
        question = pergunta,
        conversation_history = conversation_history
        )
    
    resposta = llm.complete(prompt_formated).text
    
    print('------------------------')
    print("Pergunta: ", pergunta)
    print("Resposta: ", resposta)
    print('------------------------')
    
    conversation_history.append({"role": "user", "content": pergunta})
    conversation_history.append({"role": "assistant", "content": resposta})
    
```
